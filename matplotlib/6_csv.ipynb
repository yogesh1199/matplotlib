```python
import matplotlib.pyplot as plt
import csv
```

```python
x=[]
y=[]
with open('p.csv','r') as cs_file:
    plots=csv.reader(cs_file)
    
    for i in plots:
        x.append(i[0])
        y.append(i[1])
```

```python
plt.plot(x,y,label='file')
plt.title('file')
plt.legend()
plt.show()
```
