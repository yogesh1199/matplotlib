```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5,6]
y=[11,22,33,44,55,66]

tick_label=["one","two","three","four","five","six"]
plt.bar(x,y,tick_label=tick_label,color=["red","blue","green"])
# tick_label tells particular bar name
plt.title('bargraph')
```

```python
a=[5,2,4,6,2,1,5,2,46,2,2,5,4,8,5,5,5,5,5,56,5,5,5]
range=(0,100)
b=10

plt.hist(a,b,range,histtype='bar',rwidth=0.7)
plt.title('histograph')
plt.show()
```

```python
plt.scatter(x,y,label='star',marker='*',s=50)
plt.title('scatter')
plt.legend() # show label
plt.show()
```

```python
act=['class','home','tution','pg']
slicees=[5,6,2,3]
color=['r','b','y','m']

plt.pie(slicees,labels=act,colors=color,startangle=90,shadow=True,autopct='%1.0f%%',explode=(0.3,0,0,0))
# starangle tells the starting angle
# autopct tells th percentage value
# explode makes space for particual pie 
plt.legend()
plt.show()
```
