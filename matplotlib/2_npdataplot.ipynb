```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0.0,2.0,0.01)
y=2+ np.cos(2*np.pi*x)
plt.plot(x,y,'g')
plt.grid()
plt.title('cos')
```

```python
x=np.arange(0.0,2.0,0.01)
y=2+ np.sin(2*np.pi*x)
plt.plot(x,y,'*')
plt.grid()
plt.title('sin')
```

```python
x=np.arange(0.0,2.0,0.01)
y=2+ np.tan(2*np.pi*x)
plt.plot(x,y,'-')
plt.grid()
plt.title('tan')
```
